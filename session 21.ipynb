<h3> 1.Create a Python class called Product with a method get_discount() that returns 0. Write a subclass called Electronics that overrides get_discount() to return 10.

```python
class Product:
    def get_discount(self):
        return 0

class Electronics(Product):
    def get_discount(self):
        return 10


generic_item = Product()
laptop = Electronics()

print(f"Discount on general product : {generic_item.get_discount()}%")
print(f"Discount on electronics     : {laptop.get_discount()}%")
```

<h3> 2. Build a class FoodOrder with a method calculate_total() that returns the base price. Create a subclass ZomatoOrder that overrides calculate_total() to add a 5% delivery charge.


```python
class FoodOrder:
    def __init__(self, base_price):
        self.base_price = base_price
        
    def calculate_total(self):
        return self.base_price

class ZomatoOrder(FoodOrder):
    def calculate_total(self):
        original_price = super().calculate_total()
        delivery_charge = original_price * 0.05
        
        return original_price + delivery_charge

dine_in_order = FoodOrder(500)
zomato_delivery = ZomatoOrder(500)

print(f"Dine-in Total : ₹{dine_in_order.calculate_total()}")
print(f"Zomato Total  : ₹{zomato_delivery.calculate_total()}")
```

<h3>3.Write a function show_bonus(employee) that takes any object with a bonus() method and prints the result. Test it with two classes: Influencer (bonus returns 2000) and BrandManager (bonus returns 5000), demonstrating polymorphism.


```python
class Influencer:
    def bonus(self):
        return 2000

class BrandManager:
    def bonus(self):
        return 5000

def show_bonus(employee):
    bonus_amount = employee.bonus()
    print(f"Bonus Processed: ₹{bonus_amount}")


rohan_influencer = Influencer()
priya_manager = BrandManager()

print("Testing with Influencer:")
show_bonus(rohan_influencer)
print("Testing with Brand Manager:")
show_bonus(priya_manager)
```

<h3> 4.Given this code: class User: def get_status(self): return 'active' class PremiumUser(User): pass. Update PremiumUser to override get_status() so it returns 'premium'. Then, create one User and one PremiumUser and print their statuses.<br><br><em><strong>Hint:</strong> Use the same method name in both classes to override.</em>

```python
class User:
    def get_status(self):
        return 'active'

class PremiumUser(User):
    def get_status(self):
        return 'premium'


normal_user = User()
pro_user = PremiumUser()

print(f"Normal User Status  : {normal_user.get_status()}")
print(f"Premium User Status : {pro_user.get_status()}")
```
