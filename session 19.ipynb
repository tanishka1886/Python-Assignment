<h3> 1.Define a Python class called Song with attributes title, artist, and duration (in seconds), and use the __init__() constructor to initialize these values when creating an object.

```python
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration
    
    def get_details(self):
        return f"{self.title}' by {self.artist} ({self.duration} sec)"


song1 = Song("Palti Fire", "Sambata", 268)
song2 = Song("Nusta Paisa", "MCstan", 268)
print(song1.get_details())
print(song2.get_details())
```

<h3> 4.Create a class called FoodOrder with attributes restaurant_name, items (a list), and total_price. Add a method add_item(self, item, price) that adds the item to the items list and updates total_price. Demonstrate by creating a FoodOrder object and adding two items like you would on Zomato.


```python
class FoodOrder:
    def __init__(self, restaurant_name):
        self.restaurant_name = restaurant_name
        self.items = []     
        self.total_price = 0 
        
    def add_item(self, item, price):
        self.items.append(item)       
        self.total_price += price     
        print(f"✅ Added 1x {item} (₹{price})")

my_order = FoodOrder("Burger King")
print(f"Ordering from: {my_order.restaurant_name}")

my_order.add_item("Crispy Veg Burger", 70)
my_order.add_item("Peri Peri Fries", 109)
my_order.add_item("Cold Coffee", 150)

print(f"Final Cart Items : {my_order.items}")
print(f"Total Bill       : ₹{my_order.total_price}")
```

<h3> 5.Refactor your Song class so that it also tracks a play_count attribute (starting at 0), and add a method increment_play_count(self) that increases play_count by 1 each time it's called. Show how you would use this to count how many times a user plays a song.<br><br><em><strong>Hint:</strong> Call increment_play_count() multiple times and print play_count to see the update.</em>

```python
class Song:
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration
        self.play_count = 0 
        
    def play_preview(self):
        print(f"Playing preview of '{self.title}'...")
        self.increment_play_count()

    def increment_play_count(self):
        self.play_count += 1
        print(f"(Stats: '{self.title}' has been played {self.play_count} times)")


my_song = Song("Coolie", "Gangsta's Paradise", 268)
my_song.play_preview() 
my_song.play_preview() 
my_song.play_preview() 
print(f"Total plays for '{my_song.title}': {my_song.play_count}")
```
