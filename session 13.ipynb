### 1. Write a recursive function in Python called reverse_string(s) that takes a string and returns it reversed (e.g., 'hello' becomes 'olleh').

```python
def reverse_string(s):
    if s == "":
        return ""

    return reverse_string(s[1:]) + s[0]

print(reverse_string("hello"))
```

### 2.Build a recursive function sum_playlist_durations(durations) that takes a list of song durations (in seconds) and returns the total duration, similar to how Spotify totals a playlist.

```python
def sum_playlist_durations(durations):
    
    if durations == []:
        return 0

    return durations[0] + sum_playlist_durations(durations[1:])


playlist = [180, 240, 150, 300]
print(sum_playlist_durations(playlist))
```

### 3.Given the following code, identify whether the variable 'count' is local or global in each function, and explain what will be printed when run
### count = 10 def update_count(): count = 5 print('Inside:', count)
### update_count() print('Outside:', count)

```python
count = 10   # Global variable because it is declared outside the function

def update_count():
    count = 5   # Local variable because it is declared inside the function
    print('Inside:', count)   # Prints the local value (5)

update_count()   # Calls the function
print('Outside:', count)   # Prints the global value (10)
```

### 4.Create a recursive function count_likes(posts) that takes a nested dictionary representing Instagram posts and their replies (each with a 'likes' key), and returns the total number of likes across all posts and replies.


```python
def count_likes(posts):
    total = posts["likes"]  # Likes of current post

    # Check if there are replies
    if "replies" in posts:
        for reply in posts["replies"]:
            total += count_likes(reply)  # Recursive call

    return total


# Example
posts = {
    "likes": 100,
    "replies": [
        {
            "likes": 30,
            "replies": [
                {"likes": 10},
                {"likes": 20}
            ]
        },
        {
            "likes": 40,
            "replies": [
                {"likes": 15}
            ]
        }
    ]
}

print("Total Likes:", count_likes(posts))
```

### 5.Write a Python script that demonstrates the lifetime of a local variable inside a function versus a global variable by printing their values before, during, and after a function call. Use variable names similar to 'user_status' and 'app_status', inspired by WhatsApp online/offline status.

```python
app_status = "Online"   # Global variable

print("Before function call:")
print("App Status:", app_status)


def show_status():
    user_status = "Typing..."   # Local variable
    print("\nInside function:")
    print("User Status:", user_status)
    print("App Status:", app_status)


show_status()

print("\nAfter function call:")
print("App Status:", app_status)
```
