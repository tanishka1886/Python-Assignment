Write a Python script that intentionally mixes tabs and spaces for indentation, then fix the script so it runs without errors.<br><br><em><strong>Hint:</strong> Use only spaces for indentation, as per Python's best practices.</em>

```python
if True:
	print("Hello")      # Indented using a tab
    print("World")     # Indented using spaces
```

```python
if True:
    print("Hello")
    print("World")
```

Explaination : Python uses indentation to define blocks of code. Mixing tabs and spaces for indentation can cause a TabError. To avoid this error, Python's best practice is to use 4 spaces for each indentation level consistently throughout the program.
