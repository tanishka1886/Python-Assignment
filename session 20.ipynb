### 1.Create a Python class called Product with a private attribute _price. Initialize _price in the constructor and write a method to display its value.

```python
class Product:
    def __init__(self, price):
        self._price = price   # Private attribute

    def display_price(self):
        print("Product Price:", self._price)



p1 = Product(500)

p1.display_price()
```

### 2.Add getter and setter methods for the _price attribute in your Product class to safely access and update the price. Make sure the setter prevents setting a negative price.

Hint: Raise a ValueError if the new price is less than zero.

```python
class Product:
    def __init__(self, price):
        self._price = price

    # Getter
    def get_price(self):
        return self._price

    # Setter
    def set_price(self, new_price):
        if new_price < 0:
            raise ValueError("Price cannot be negative.")
        self._price = new_price


p1 = Product(1000)

print("Current Price:", p1.get_price())


p1.set_price(1500)
print("Updated Price:", p1.get_price())


try:
    p1.set_price(-500)
except ValueError as e:
    print(e)
```

### 3.Build a class called Playlist that has a private attribute _songs (a list of song names). Write methods to add a song, remove a song, and get the current list of songs using proper encapsulation.

```python
class Playlist:
    def __init__(self):
        self._songs = []

    def add_song(self, song):
        self._songs.append(song)
        print(song, "added.")

    def remove_song(self, song):
        if song in self._songs:
            self._songs.remove(song)
            print(song, "removed.")
        else:
            print(song, "not found.")

    def get_songs(self):
        return self._songs



playlist = Playlist()

playlist.add_song("Kesariya")
playlist.add_song("Apna Bana Le")
playlist.add_song("Tum Hi Ho")

print("Playlist:", playlist.get_songs())

playlist.remove_song("Apna Bana Le")

print("Updated Playlist:", playlist.get_songs())
```

### 4.Create an abstract class PaymentMethod with an abstract method pay(amount). Then, create two subclasses: UPI and CreditCard, each implementing the pay method with a print statement showing how the payment would be processed.

Hint: Use the abc module for abstraction.

```python
from abc import ABC, abstractmethod


class PaymentMethod(ABC):

    @abstractmethod
    def pay(self, amount):
        pass



class UPI(PaymentMethod):
    def pay(self, amount):
        print(f"Paid ₹{amount} using UPI.")



class CreditCard(PaymentMethod):
    def pay(self, amount):
        print(f"Paid ₹{amount} using Credit Card.")



u = UPI()
c = CreditCard()

u.pay(500)
c.pay(1200)
```
