### 1.Use re.findall() to extract all valid Indian phone numbers (10 digits, starting with 7, 8, or 9) from a given text string that contains random numbers, prices, and phone numbers like those seen in OLX or WhatsApp chats.

```python
import re

text = """
Available items:
Laptop - ₹45000
Call me at 9876543210
Office: 7894561230
Old number: 6123456789
Price: 25000
WhatsApp: 9123456789
OTP: 456789
Emergency: 8765432109
Random number: 1234567890
"""

phone_numbers = re.findall(r'\b[789]\d{9}\b', text)

print("Valid Indian Phone Numbers:")
for number in phone_numbers:
    print(number)
```

### 2.Write a Python function using re.search() that checks if a given string contains a valid date in the format DD/MM/YYYY (e.g., 25/06/2024), and returns True if found, otherwise False.

Hint: Use the pattern '\b\d{2}/\d{2}/\d{4}\b'.

```python
import re

def contains_valid_date(text):
    pattern = r'\b\d{2}/\d{2}/\d{4}\b'

    if re.search(pattern, text):
        return True
    else:
        return False

text1 = "The meeting is on 25/06/2024."
text2 = "The meeting is on 25-06-2024."

print(contains_valid_date(text1))
print(contains_valid_date(text2))
```

### 3.Given a messy text copied from a Zomato review containing multiple emails, use re.findall() to extract all valid email addresses and print them as a list.

```python
import re

text = """
Great food and quick delivery!
For support contact support@zomato.com
You can also email raj123@gmail.com for offers.
Business queries: sales.team@company.co.in
Invalid emails:
abc@gmail
user@.com
hello@com
"""

emails = re.findall(r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}', text)

print(emails)
```

### 4.Use re.sub() to mask all but the last 4 digits of any phone number in a string (e.g., replace 9876543210 with ******3210) like Paytm does for privacy.

Constraint: Do not use loops; achieve this only with re.sub().

```python
import re

text = """
Call me at 9998279512.
WhatsApp: 8866676201
Office: 7894561230
"""

masked_text = re.sub(r'(\d{6})(\d{4})', r'******\2', text)

print(masked_text)
```

### 5.Use ChatGPT to generate a regex pattern that matches Flipkart-style order IDs (e.g., OD123456789012345000) and test it in Python using re.search() on sample order strings.

```python
import re

# Regex pattern for Flipkart-style Order ID
pattern = r'\bOD\d{18}\b'

# Sample order strings
order1 = "Your Flipkart Order ID is OD123456789012345000."
order2 = "Tracking ID: OD987654321098765432"
order3 = "Order Number: 1234567890"
order4 = "Order ID: OD12345"

# Test using re.search()
print(bool(re.search(pattern, order1)))  # True
print(bool(re.search(pattern, order2)))  # True
print(bool(re.search(pattern, order3)))  # False

print(bool(re.search(pattern, order4)))  # False
```
